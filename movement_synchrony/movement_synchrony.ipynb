```python
videos = ["RPS_C_13_C3.wmv","RPS_FP_13_C3.wmv","RPS_PS_13_C3.wmv",
        "RPS_C_15_C3.wmv","RPS_FP_15_C3.wmv","RPS_PS_15_C3.wmv",
        "RPS_C_16_C3.wmv","RPS_FP_16_C3.wmv","RPS_PS_16_C3.wmv",
        "RPS_C_17_C3.wmv","RPS_FP_17_C3.wmv","RPS_PS_17_C3.wmv"         
         ]
jsons_for_video = []
vmrk_for_video = []

for video in videos:
    jsons_for_video.append([j for j in jsons if video.split(".")[0] in j])
    
for video in videos:
    vmrk_for_video.extend([v for v in vmrks if video[0:-7] in v])

folder = "./data/"

mot_for_roi_for_video = []#will contain lists containing motion for the left and the right participant
```

```python
for vid, jsons in zip(videos,jsons_for_video):#for each video/vmrk
    ! ./decode_motion {folder+vid} {folder+"temp.mot"}
    mot_for_video = []
    for j in jsons:#different rois
        with open(folder+j, encoding='utf-8') as fh:
            roi = json.load(fh)
            h264 = tables.open_file(folder+"temp.mot", mode='r')
            h264 = h264.root.motion_tensor
            mothistmap = helper.weighted_histograms(h264,roi=roi,weight_by_pixels=True)
            
            hist = ndimage.gaussian_filter(mothistmap,10)
            amount = np.sum(hist,axis=0)#1d signal
            mot_for_video.append(amount)
    mot_for_roi_for_video.append(mot_for_video)
    ! rm {folder+"temp.mot"}
```

```python
def game_period(vmrk,len_of_mot):
    """ Parses vmrk, obtains start trigger, reads eyes open and closed triggers and returns 
        trial times as boolen array
        Args:
            vmrk: Path to vmrk file
            len_of_mot: length of motion file
        Returns:
            Game period as boolean array. True for game period.        
    """
    vmrk = helper.parse_vmrk(vmrk)
    video_start_trigger = np.array(vmrk["time"], dtype = int)[np.array(vmrk['description'])=='S128'][0]

    eyes_open = np.array(vmrk['description'])=='S 21'
    eyes_open = np.array(vmrk["time"], dtype = np.int32)[eyes_open]
    eyes_open = (eyes_open - video_start_trigger)//20

    eyes_closed = np.array(vmrk['description'])=='S 20'
    eyes_closed = np.array(vmrk["time"], dtype = np.int32)[eyes_closed]
    eyes_closed = (eyes_closed - video_start_trigger)//20

    game = np.zeros(len_of_mot,dtype = np.bool)
    game[eyes_open[0]:eyes_closed[1]] = 1
    game[eyes_open[1]:eyes_closed[2]] = 1
    return game
```

```python
poi_roi = []
for mot_for_roi, vmrk in zip(mot_for_roi_for_video, vmrk_for_video):
    mot_roi = []
    for mot in mot_for_roi:
        mask = game_period(folder+vmrk, len(mot))
        mot_roi.append(mot[mask])
    poi_roi.append(mot_roi)
    
poi_roi = np.array(poi_roi)
```

```python
cond_1 = poi_roi[::3]
cond_2 = poi_roi[1::3]
cond_3 = poi_roi[2::3]
```

```python
df = pd.DataFrame()

for condition, cond_name in zip([cond_1,cond_2,cond_3],["Condition 1", "Condition 2", "Condition 3"]):
    condition_correlations=[]
    for participants in condition:
        corr = stats.linregress(participants[0],participants[1])[2]
        condition_correlations.append(corr)
    df[cond_name] = condition_correlations
    

#https://www.researchgate.net/post/average_of_Pearson_correlation_coefficient_values

df = df.append(np.tanh(np.arctanh(df).mean(numeric_only=True)), ignore_index=True)
df["Mean"] = np.tanh(np.arctanh(df).mean(numeric_only=True, axis=1))
df["Participant"] = ["Dyad 1","Dyad 2","Dyad 3"," Dyad 4","Mean"]
df.set_index("Participant",inplace=True)
df = pd.DataFrame(df.round(2), dtype=str)
df["Mean"][-1] = ""

helper.print_table(df,"Table S2:      Correlation for dyads within condition",0,
                  "Fishers z-transformation was applied before calculating means. <br> "+
                   "See Alexander (1990). For each dyad and condition: n>12000.")
```
